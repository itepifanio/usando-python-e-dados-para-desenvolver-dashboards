```python
df_turmas = pd.read_csv('turmas-2022.2.csv', sep=';')
```

```python
df_turmas = df_turmas.drop(columns=['codigo_turma', 'matricula_docente_externo', 'observacao', 'ch_dedicada_periodo', 'nivel_ensino', 'campus_turma', 'data_inicio', 'data_fim', 'descricao_horario', 'total_solicitacoes', 'capacidade_aluno', 'distancia', 'tipo', 'data_consolidacao', 'agrupadora', 'id_turma_agrupadora', 'qtd_aulas_lancadas', 'convenio', 'modalidade_participantes', 'local'])
df_turmas.head()
```

```python
df_matriculas = df_matriculas.merge(df_turmas, how='inner', on='id_turma')
```
